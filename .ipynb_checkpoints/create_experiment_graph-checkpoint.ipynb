```python
import plotly
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
import numpy as np

init_notebook_mode(connected=True)
```

```python
def get_acc_list(filename):
    csv = np.genfromtxt(filename, delimiter="|", skip_header=1)
    
    epoch, acc, loss, lr, val_acc, val_loss = np.hsplit(csv, 6)
    
    val_acc = val_acc * 100
    val_acc = val_acc.flatten()
    return val_acc
```

```python
experiment = "2"
epoch = np.arange(1120)

names = ["log_"+ experiment +"-1", "log_"+ experiment +"-2", "log_"+ experiment +"-3", "log_"+ experiment +"-4"]
all_list = []
folder = "./log_file/"
extension = ".csv"

for name in names:
    filename = folder + name + extension
    acc_list = get_acc_list(filename)
    all_list.append(acc_list)
```

```python
title = "Validation Accuracy " + str(names)
fig_name = experiment + "_acc.html"

trace1 = go.Scatter(
    x = epoch,
    y = all_list[0],
    mode = 'lines',
    name = 'Train: Base'
)

trace2 = go.Scatter(
    x = epoch,
    y = all_list[1],
    mode = 'lines',
    name = 'Train: Base + 3Model'
)

trace3 = go.Scatter(
    x = epoch,
    y = all_list[2],
    mode = 'lines',
    name = 'Train: Base + PCAJ'
)

trace4 = go.Scatter(
    x = epoch,
    y = all_list[3],
    mode = 'lines',
    name = 'Train: Base + 3Model + PCAJ'
)

layout = go.Layout(
    width=1000,
    height=500,
    title= title
)
data = [trace1, trace2, trace3, trace4]
fig1 = go.Figure(data=data, layout=layout)
plotly.offline.iplot(fig1)
```

```python
plotly.offline.plot(fig1, filename=fig_name)
```
