```python
import plotly
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
import numpy as np

init_notebook_mode(connected=True)
```

---
# Read CSV Training File

```python
name = "log_1-2"
extension = ".csv"

csv = np.genfromtxt(name + extension, delimiter="|", skip_header=1)
print(csv)
```

```python
print(type(csv))
print(len(csv[0]))
```

```python
# epoch | acc | loss | lr | val_acc | val_loss
epoch, acc, loss, lr, val_acc, val_loss = np.hsplit(csv, 6)
```

```python
acc = acc * 100
val_acc = val_acc * 100

acc = acc.flatten() 
#print(acc)

val_acc = val_acc.flatten()
#print(val_acc)

loss = loss.flatten()

val_loss = val_loss.flatten()

epoch = epoch.flatten()
#print(epoch)
```

```python
title = "Training Accuracy vs Validation Accuracy"
fig_name = name + "_acc.html"

trace1 = go.Scatter(
    x = epoch,
    y = acc,
    mode = 'lines',
    name = 'Train'
)

trace2 = go.Scatter(
    x = epoch,
    y = val_acc,
    mode = 'lines',
    name = 'Validation'
)

layout = go.Layout(
    width=800,
    height=500,
    title= title
)
data = [trace1, trace2]
fig1 = go.Figure(data=data, layout=layout)
plotly.offline.iplot(fig1)
```

```python
plotly.offline.plot(fig1, filename=fig_name)
```

```python
title = "Training Loss vs Validation Loss"
fig_name = name + "_loss.html"

trace1 = go.Scatter(
    x = epoch,
    y = loss,
    mode = 'lines',
    name = 'Train'
)

trace2 = go.Scatter(
    x = epoch,
    y = val_loss,
    mode = 'lines',
    name = 'Validation'
)

layout = go.Layout(
    width=800,
    height=500,
    title= title
)
data = [trace1, trace2]
fig2 = go.Figure(data=data, layout=layout)
plotly.offline.iplot(fig2)
```

```python
plotly.offline.plot(fig2, filename=fig_name)
```
